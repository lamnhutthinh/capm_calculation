# Câu 1

```python
df=pd.read_csv(r'C:\Users\Khang Lam\Downloads\capm-data.csv',index_col=0,parse_dates=True)
df
```

```python
df=df.dropna()
df=df.resample('M').last()
df
```

```python
df.reset_index(inplace=True)
```

```python
df['Date']=df['Date'].astype('str')
df['Date']=df['Date'].apply(lambda x: x.split('-')[1]+'/'+x.split('-')[0])
df
```

```python
df.set_index('Date',inplace=True)
df
```

```python
df_r=df.pct_change()
df_r=df_r.dropna()
df_r.columns=['VNINDEX Rm','Vietcombank Ri','FPT Ri']
df_r
```

```python
df_cov=df_r.cov()
df_cov
```

```python
beta_Vietcom=df_cov.iloc[0,1]/(df_r['VNINDEX Rm'].var())
print('Beta của Vietcombank ở trường hợp câu 1 là:',beta_Vietcom)
beta_FPT=df_cov.iloc[0,2]/(df_r['VNINDEX Rm'].var())
print('Beta của FPT ở trường hợp câu 1 là:',beta_FPT)
```
